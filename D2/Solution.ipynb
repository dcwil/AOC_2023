```python
import re
import pandas as pd
```

```python
with open('data.txt', 'r') as h:
    example = h.read()
```

```python
pattern = r'\n'


rows_ls = []
for line in re.split(pattern, example):
    game_number, games = re.split(r':', line)
    game_number = int(game_number.strip('Game '))


    colour_patterns = {
        'red': r'\d+ red',
        'blue': r'\d+ blue',
        'green': r'\d+ green',
    }
    
    for set_i, cube_set in enumerate(games.split(';')):
        row = {'game': game_number, 'set': set_i}
        for colour, pattern in colour_patterns.items():
            cubes = re.findall(pattern, cube_set)

            if len(cubes) != 0:
                row[colour] = int(cubes[0].rstrip(f' {colour}'))
            else:
                row[colour] = 0
        rows_ls.append(row)
            
            
        
```

```python
df = pd.DataFrame(rows_ls)
df
```

```python
limits = (df['red'] <= 12) & (df['green'] <= 13) & (df['blue'] <= 14)
excluded_games = set(df[~limits]['game'])
all_games = set(df['game'])
```

```python
sum(all_games - excluded_games)
```

```python
max_df = df.groupby('game').max()
max_df['power'] = max_df['red'] * max_df['blue'] * max_df['green']
max_df
```
